```python
import pandas as pd
```

```python
# datasetpath
my_path = "../datasets/deliveries.csv"
```

# **Export to CSV**


```python
df = pd.DataFrame()
# df.to
# this to is used to export to muliple format
```

```python
df = pd.read_csv(my_path)
df.head()
```

```python
# now suppose we need a new dataset which contains a batsman name and his total runs

# df.groupby('batsman')['total_runs'].sum()


# for name and runs
temp_df = (
    df.groupby("batsman")["total_runs"].sum().sort_values(ascending=False).reset_index()
)
temp_df
```

```python
# export this to csv
# Index false because we dont need extra column in csv because when we load it again we automatically gets the index by pandas

temp_df.to_csv("../datasets/my_custom_batsman_runs.csv", index=False)
```

```python
# batsman vs team total runs

temp_df = df.pivot_table(
    index="batsman", columns="bowling_team", values="batsman_runs", aggfunc="sum"
)
temp_df
```

```python
temp_df.to_csv("../datasets/batsman_vs_teams.csv")
```

# **Export to HTML**

- we can convert our analysis as html table
- usefull when we need to convert our analysis as a blog to write somewhere


```python
# har over ki konsi ball pe kitne runs pade hai aaj tak

temp_df = df.pivot_table(
    index="over", columns="ball", values="batsman_runs", aggfunc="sum"
)
temp_df
```

```python
# har over ki konsi ball pe kitne 6 pade hai aaj tak
# run pivot table on values where batsmann run == 6
# either we can filter or directly apply query


temp_df = df.query("batsman_runs == 6").pivot_table(
    index="over", columns="ball", values="batsman_runs", aggfunc="count"
)
temp_df
```

```python
temp_df.to_html("../datasets/sixes.html")
```

# **Export to JSON**


```python
# lets create a json as batsman name, there team name and runs made

temp_df = df.groupby(["batting_team", "batsman"])["total_runs"].sum().unstack()

temp_df
```

```python
temp_df.to_json("../datasets/ipl_runs.json")
```
